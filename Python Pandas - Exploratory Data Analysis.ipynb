# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ADMIN\OneDrive - siomairhys\Portfolio Project\Worl Population Dataset.csv")
df
```

```python
# Setting the numbers in two decimal place
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Looking at the info and watching the values ib every columns
df.info()
```

```python
# Getting the overview of each population
df.describe()
```

```python
# Display the missing datas
df.isnull().sum()
```

```python
df.nunique()
```

```python
# Ordering/Sorting the datas 
df.sort_values(by = "2022 Population", ascending = False).head(10)
df.sort_values(by = "2015 Population", ascending = False).head(10)
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

```python
# Viewing correlation in the datas
df.corr()
```

```python
# Visualizing the correlation
sns.heatmap(df.corr(), annot = True)

plt.rcParams['figure.figsize'] = (15,10)

plt.show()
```

```python
# Grouping by the datas and diving more in the highest population
df.groupby('Continent').mean().sort_values(by = "2022 Population", ascending = False)
```

```python
df[df['Continent'].str.contains('Asia')]
```

```python
df2 = df.groupby('Continent').mean().sort_values(by = "2022 Population", ascending = False)
df2
```

```python
df2.transpose()
```

```python
df2.plot()
```

```python
# Ordering the columns to 1970 - 2022 populations for better visualization later
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
# Now transposing it for visualization
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
