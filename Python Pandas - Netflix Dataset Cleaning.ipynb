# Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ADMIN\OneDrive - siomairhys\Portfolio Project\Resources,Templates and Codes\Netflix Userbase Data Cleaning Panda.csv")
df
```

```python
# I have removed one duplicate in the dataset
df = df.drop_duplicates()
df
```

```python
# Removing the two columns in the dataset
df.drop(["Device", "Plan Duration"], axis = 1, inplace = True)
df
```

```python
# Remove the unneccessary values in the datasets
df["Subscription Type"] = df["Subscription Type"].str.lstrip("/")
df["Subscription Type"] = df["Subscription Type"].str.rstrip("-")
df["Subscription Type"] = df["Subscription Type"].str.rstrip("12")
df["Subscription Type"] = df["Subscription Type"].str.rstrip("_")
df["Subscription Type"] = df["Subscription Type"].str.rstrip("1")
df["Subscription Type"] = df["Subscription Type"].str.rstrip("31")

# Or you can use strip only to get rid of the unneccessary datas
df["Subscription Type"] = df["Subscription Type"].str.strip("/-12_131")

df
```

```python
# Replacing the underscore sign in the Basic Subscription type
df["Subscription Type"] = df["Subscription Type"].str.replace("_", "")
df
```

```python
# Cleaning the Join Date columns and fixing the dates
df["Join Date"] = pd.to_datetime(df['Join Date'], errors='coerce').dt.strftime('%d-%m-%y')
df
```

```python
# Cleaning the Last Payment Date columns and fixing the dates
df["Last Payment Date"] = pd.to_datetime(df['Last Payment Date'], errors='coerce').dt.strftime('%d-%m-%y')
df
```

```python
# Fixing and Clenaing the column names
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('-', '_').str.replace('/', '_')
df
```

```python
# Cleaning the Country names
df['Country'] = df['Country'].str.strip().str.title()
df
```

```python
# Saving the cleaned dataset to my laptop and making a new csv file
df.to_csv(r'C:\Users\ADMIN\OneDrive - siomairhys\Portfolio Project\Netflix Userbase Data Cleaned Panda.csv', index = False)
```
